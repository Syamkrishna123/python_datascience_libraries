# histogram


```python
data = pd.read_csv('auto-mpg.csv')
data.head()
```

```python
plt.figure()
plt.hist(data['mpg'])
plt.title('MPG PLOTING')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'])
plt.title('MPG PLOTING')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'])
plt.title('distribution of mpg',fontsize=30)
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red')
plt.title('distribution of mpg',fontsize=30)
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red')
plt.title('distribution of mpg',fontsize=30)
plt.xlabel('MPG')
plt.ylabel('FREQUENCY')
plt.xticks()
plt.yticks()

```

```python
plt.figure(figsize=(8,8),edgecolor='green',facecolor='blue')
plt.hist(data['mpg'],color='red',rwidth=0.7)
plt.title('MPG',fontsize=30)
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.xticks()
```

```python
plt.figure(figsize=(8,8),edgecolor='green',facecolor='blue')
plt.hist(data['mpg'],color='red',rwidth=0.7,density=True)
plt.title('MPG',fontsize=30)
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.xticks()
plt.yticks()
```

```python
plt.figure(figsize=(8,8))
plt.hist(data['mpg'],color='red',rwidth=0.7,density=True,cumulative=True)
plt.title('MPG',fontsize=30)
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.xticks()
plt.yticks()
```

# scatter plot


```python
plt.figure(figsize=(8,8))
plt.scatter(data1['horsepower'],data1['mpg'])
plt.title('horsepower vs mpg',fontsize=30)
plt.xlabel('HP')
plt.ylabel('MPG')
```

```python
#Changing the color of the graph and marker
plt.figure(figsize=(8,8))
plt.scatter(data1['horsepower'],data1['mpg'],c='red',marker='*')
plt.title('HORSEPOWER Vs MPG',fontsize=20)
plt.xlabel('HP')
plt.ylabel('MPG')
```

```python
data1
```

```python
plt.figure(figsize=(8,8))
plt.scatter(data1['displacement'],data1['mpg'],c='red',marker='*')
plt.title('DISPLACEMENT Vs MPG',fontsize=20)
plt.xlabel('Displacement')
plt.ylabel('Mpg')
```

```python
plt.figure(figsize=(8,8))

plt.subplot(1,2,1)
plt.scatter(data1['horsepower'],data1['mpg'],c='red',marker='*')
plt.title('HORSEPOWER Vs MPG',fontsize=20)
plt.xlabel('HP')
plt.ylabel('MPG')

plt.subplot(1,2,2)
plt.scatter(data1['displacement'],data1['mpg'],c='red',marker='*')
plt.title('DISPLACEMENT Vs MPG',fontsize=20)
plt.xlabel('Displacement')
plt.ylabel('Mpg')
```
